```python
import numpy as np
```

```python
# Tensor Simples 2x2x3
# Formato: Profundidade/Canais, Linhas, Colunas
# np.zeros(inicializa tudo com 0)
imagem_rgb = np.zeros((3,2,2))
imagem_rgb[0, : , :] = 255
print("Imagem RGB (Apenas vermelho preenchido)")
print(imagem_rgb)
```

```python
from matplotlib import pyplot as plt
# Pyplot Transpose = Inverte a ordem para Linha, Coluna, Canais
plt.imshow(imagem_rgb.transpose(1, 2, 0).astype(np.uint8),
           vmin=0,vmax=255)
plt.title("Vermelho")
plt.show()
```

```python
# Quadrado amarelo = r=255,g=255,b=0
imagem_rgb[0, : , :] = 255
imagem_rgb[1, : , :] = 255
imagem_rgb[2, : , :] = 0
plt.imshow(imagem_rgb.transpose(1, 2, 0).astype(np.uint8),
           vmin=0,vmax=255)
plt.title("Amarelo")
plt.show()
```

```python
# Quadrado ciano = r=0,g=255,b=255
imagem_rgb[0, : , :] = 0
imagem_rgb[1, : , :] = 255
imagem_rgb[2, : , :] = 255
plt.imshow(imagem_rgb.transpose(1, 2, 0).astype(np.uint8),
           vmin=0,vmax=255)
plt.title("Ciano")
plt.show()
```

```python
# Quadrado branco = r=255,g=255,b=255
imagem_rgb[0, : , :] = 255
imagem_rgb[1, : , :] = 255
imagem_rgb[2, : , :] = 255
plt.imshow(imagem_rgb.transpose(1, 2, 0).astype(np.uint8),
           vmin=0,vmax=255)
plt.title("Branco")
plt.show()
```

```python
# Quadrado preto = r=0,g=0,b=0
imagem_rgb[0, : , :] = 0
imagem_rgb[1, : , :] = 0
imagem_rgb[2, : , :] = 0
plt.imshow(imagem_rgb.transpose(1, 2, 0).astype(np.uint8),
           vmin=0,vmax=255)
plt.title("Preto")
plt.show()
```
